```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

# Historical Data
years = list(range(1999, 2024))
births = [134.0, 133.7, 133.5, 133.3, 133.2, 133.1, 133.0, 132.9, 132.8, 132.7,
          132.6, 132.5, 132.4, 132.3, 132.2, 132.1, 132.0, 131.9, 131.8, 131.7,
          131.6, 131.5, 131.4, 131.3, 131.2]
deaths = [52.6, 52.9, 53.2, 53.5, 53.8, 54.0, 54.3, 54.6, 54.9, 55.2,
          55.5, 55.8, 56.1, 56.4, 56.7, 57.0, 57.3, 57.6, 57.9, 58.2,
          58.5, 58.8, 59.1, 59.4, 59.7]

# Convert to NumPy arrays for calculations
years_array = np.array(years)
births_array = np.array(births)
deaths_array = np.array(deaths)

# Fit linear regression models
slope_births, intercept_births, _, _, _ = linregress(years_array, births_array)
slope_deaths, intercept_deaths, _, _, _ = linregress(years_array, deaths_array)

# Function to predict births and deaths
def predict_births(year):
    return slope_births * year + intercept_births

def predict_deaths(year):
    return slope_deaths * year + intercept_deaths

# Project into the future until the two trends intersect
future_years = []
projected_births = []
projected_deaths = []

current_year = years[-1] + 1

while True:
    birth_prediction = predict_births(current_year)
    death_prediction = predict_deaths(current_year)

    # Append the projections
    future_years.append(current_year)
    projected_births.append(birth_prediction)
    projected_deaths.append(death_prediction)

    # Check if deaths surpass births
    if death_prediction >= birth_prediction:
        intersection_year = current_year
        intersection_births = birth_prediction
        intersection_deaths = death_prediction
        break
    current_year += 1

# Combine historical and projected data
all_years = years + future_years
all_births = births + projected_births
all_deaths = deaths + projected_deaths

# Plotting
plt.figure(figsize=(12, 6))

# Plot historical data
plt.plot(years, births, 'bo-', label='Historical Births')
plt.plot(years, deaths, 'ro-', label='Historical Deaths')

# Plot projected data
plt.plot(future_years, projected_births, 'b--', label='Projected Births')
plt.plot(future_years, projected_deaths, 'r--', label='Projected Deaths')

# Highlight the intersection point
plt.plot(intersection_year, intersection_births, 'go', label='Intersection Point')

# Annotate the intersection point
plt.annotate(f'Intersection in {intersection_year}',
             xy=(intersection_year, intersection_births),
             xytext=(intersection_year-10, intersection_births+5),
             arrowprops=dict(facecolor='green', shrink=0.05))

plt.title('Global Births and Deaths (1999 - Projected Intersection)')
plt.xlabel('Year')
plt.ylabel('Number (Millions)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print the year of intersection
print(f"The projected year when deaths will equal births is {intersection_year}.")
```
