```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Number of Professionals': [50000, 49000, 48000, 47000, 46000, 45000, 44000, 43000, 42000, 41000, 40000],  # Example numbers
    'Job Openings': [1000, 1200, 1500, 1800, 2200, 2500, 3000, 3500, 4000, 4500, 5000],  # Example numbers
    'Annual Pay': [50000, 51000, 52500, 53500, 54500, 55500, 56000, 57000, 58000, 59000, 60000]  # Example numbers
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(14, 7))

# Plot for Number of Professionals
plt.subplot(2, 2, 1)
sns.lineplot(x='Year', y='Number of Professionals', data=df)
plt.title('Decline in Skilled Trades Professionals')
plt.xlabel('Year')
plt.ylabel('Number of Professionals')

# Plot for Job Openings
plt.subplot(2, 2, 2)
sns.lineplot(x='Year', y='Job Openings', data=df)
plt.title('Increase in Demand for Skilled Trades')
plt.xlabel('Year')
plt.ylabel('Job Openings')

# Plot for Annual Pay
plt.subplot(2, 2, 3)
sns.lineplot(x='Year', y='Annual Pay', data=df)
plt.title('Annual Pay of Skilled Trades Professionals')
plt.xlabel('Year')
plt.ylabel('Annual Pay ($)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Assuming you have the data in a CSV file or you've entered it manually as shown before
# For this example, we'll use the manually entered data:

data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Number of Professionals': [50000, 49000, 48000, 47000, 46000, 45000, 44000, 43000, 42000, 41000, 40000],
    'Job Openings': [1000, 1200, 1500, 1800, 2200, 2500, 3000, 3500, 4000, 4500, 5000],
    'Annual Pay': [50000, 51000, 52500, 53500, 54500, 55500, 56000, 57000, 58000, 59000, 60000]
}

df = pd.DataFrame(data)

# Create the plot
fig, ax1 = plt.subplots(figsize=(14, 10))

# Plot for Number of Professionals
color = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Professionals', color=color)
ax1.plot(df['Year'], df['Number of Professionals'], color=color, label='Prof Decline')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

# Plot for Job Openings
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Job Openings', color=color)
ax2.plot(df['Year'], df['Job Openings'], color=color, label='Job Demand', linestyle='--')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

# Plot for Annual Pay
ax3 = ax1.twinx()
ax3.spines["right"].set_position(("axes", 1.1))
color = 'tab:green'
ax3.set_ylabel('Annual Pay ($)', color=color)
ax3.plot(df['Year'], df['Annual Pay'], color=color, label='Annual Pay', linestyle='-.')
ax3.tick_params(axis='y', labelcolor=color)
ax3.legend(loc='lower right')

# Title and Layout
plt.title('Trends in Skilled Trades: Professionals, Demand, and Pay Since 2014', fontsize=16)
fig.tight_layout()

# Adjusting the layout to prevent overlap
plt.subplots_adjust(right=0.8)
plt.show()
```
